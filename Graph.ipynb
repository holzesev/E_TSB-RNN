```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def print_data_test_acc(data,y_min):

    test_loss = np.genfromtxt('./datasets/' + data + '/M0_DiverSet_results/p12_test_loss20.csv', dtype=None, delimiter=',', skip_header=0)
    test_loss1 = np.genfromtxt('./datasets/' + data + '/M1_DiverSet_results/p12_test_loss20.csv', dtype=None, delimiter=',', skip_header=0)
    train_loss = np.genfromtxt('./datasets/' + data + '/M0_DiverSet_results/p12_train_loss20.csv', dtype=None, delimiter=',', skip_header=0)
    train_loss1 = np.genfromtxt('./datasets/' + data + '/M1_DiverSet_results/p12_train_loss20.csv', dtype=None, delimiter=',', skip_header=0)
    test_acc = np.genfromtxt('./datasets/' + data + '/M0_DiverSet_results/p12_test_accuracy20.csv', dtype=None, delimiter=',', skip_header=0)
    test_acc1 = np.genfromtxt('./datasets/' + data + '/M1_DiverSet_results/p12_test_accuracy20.csv', dtype=None, delimiter=',', skip_header=0)
    train_acc = np.genfromtxt('./datasets/' + data + '/M0_DiverSet_results/p12_train_accuracy20.csv', dtype=None, delimiter=',', skip_header=0)
    train_acc1 = np.genfromtxt('./datasets/' + data + '/M1_DiverSet_results/p12_train_accuracy20.csv', dtype=None, delimiter=',', skip_header=0)

    ci = 1.96*test_acc.std(1)
    ci1 = 1.96*test_acc1.std(1)
    x=range(1,len(ci)+1)
    plt.plot(x,test_acc.mean(1), label='TSB-RNN', color='r', alpha=.5)
    plt.plot(x,test_acc1.mean(1), label='ETSB-RNN', color='b', alpha=.5)
    plt.ylim(y_min-(1-y_min)/20,1+(1-y_min)/20)
    plt.xlim(-4,124)
    test_acc_mean_up=test_acc.mean(1)+ci
    test_acc_mean_up[test_acc_mean_up>1]=1
    test_acc1_mean_up=test_acc1.mean(1)+ci1
    test_acc1_mean_up[test_acc1_mean_up>1]=1
    plt.fill_between(x, (test_acc.mean(1)-ci), (test_acc_mean_up), color='r', alpha=.1)
    plt.fill_between(x, (test_acc1.mean(1)-ci1), (test_acc1_mean_up), color='b', alpha=.1)
    plt.xlabel('Number of Epochs')
    plt.ylabel('Accuracy')
    plt.xticks([0,20,40,60,80,100,120])

    current_values = plt.gca().get_yticks()
    plt.gca().set_yticklabels(['{:,.2f}'.format(x) for x in current_values])
    plt.title(data,loc='center',pad=10, fontsize=14)
    plt.legend(loc=4)
    plt.grid(True,ls='--')

    result = np.where(train_loss == np.amin(train_loss, axis=0))
    zipped = list(zip(result[0], result[1]))
    for cord in zipped:
        plt.plot(cord[0]+1, test_acc[cord], 'ro')

    result1 = np.where(train_loss1 == np.amin(train_loss1, axis=0))
    zipped1 = list(zip(result1[0], result1[1]))
    for cord in zipped1:
        plt.plot(cord[0]+1, test_acc1[cord], 'b^')
```

```python
fig = plt.figure(figsize=(12,6))
plt.subplot(2,3,1)
print_data_test_acc('Beers',0.9)

plt.subplot(2,3,2)
print_data_test_acc('Flights',0.6)

plt.subplot(2,3,3)
print_data_test_acc('Hospital',0.6)

plt.subplot(2,3,4)
print_data_test_acc('Movies',0.9)

plt.subplot(2,3,5)
print_data_test_acc('Rayyan',0.8)

plt.subplot(2,3,6)
print_data_test_acc('Tax',0.8)

plt.subplots_adjust(hspace=0.5,wspace=0.3)

fig.savefig('images/test_acc.png', dpi=100, bbox_inches='tight')
```

```python
def print_data_M1(data,y_min):

    test_loss = np.genfromtxt('./datasets/' + data + '/M0_DiverSet_results/p12_test_loss20.csv', dtype=None, delimiter=',', skip_header=0)
    test_loss1 = np.genfromtxt('./datasets/' + data + '/M1_DiverSet_results/p12_test_loss20.csv', dtype=None, delimiter=',', skip_header=0)
    train_loss = np.genfromtxt('./datasets/' + data + '/M0_DiverSet_results/p12_train_loss20.csv', dtype=None, delimiter=',', skip_header=0)
    train_loss1 = np.genfromtxt('./datasets/' + data + '/M1_DiverSet_results/p12_train_loss20.csv', dtype=None, delimiter=',', skip_header=0)
    test_acc = np.genfromtxt('./datasets/' + data + '/M0_DiverSet_results/p12_test_accuracy20.csv', dtype=None, delimiter=',', skip_header=0)
    test_acc1 = np.genfromtxt('./datasets/' + data + '/M1_DiverSet_results/p12_test_accuracy20.csv', dtype=None, delimiter=',', skip_header=0)
    train_acc = np.genfromtxt('./datasets/' + data + '/M0_DiverSet_results/p12_train_accuracy20.csv', dtype=None, delimiter=',', skip_header=0)
    train_acc1 = np.genfromtxt('./datasets/' + data + '/M1_DiverSet_results/p12_train_accuracy20.csv', dtype=None, delimiter=',', skip_header=0)

    train_ci1 = 1.96*train_acc1.std(1)
    test_ci1 = 1.96*test_acc1.std(1)
    x=range(1,len(train_ci1)+1)
    plt.plot(x,train_acc1.mean(1), ls='--', label='Train', color='g', alpha=.5)
    plt.plot(x,test_acc1.mean(1), label='Test', color='b', alpha=.5)

    plt.ylim(y_min-(1-y_min)/20,1+(1-y_min)/20)
    plt.xlim(-4,124)
    train_acc1_mean_up=train_acc1.mean(1)+train_ci1
    train_acc1_mean_up[train_acc1_mean_up>1]=1
    test_acc1_mean_up=test_acc1.mean(1)+test_ci1
    test_acc1_mean_up[test_acc1_mean_up>1]=1
    plt.fill_between(x, (train_acc1.mean(1)-train_ci1), (train_acc1_mean_up), color='r', alpha=.1)
    plt.fill_between(x, (test_acc1.mean(1)-test_ci1), (test_acc1_mean_up), color='b', alpha=.1)
    plt.xlabel('Number of Epochs')
    plt.ylabel('Accuracy')
    plt.xticks([0,20,40,60,80,100,120])
    
    current_values = plt.gca().get_yticks()
    plt.gca().set_yticklabels(['{:,.2f}'.format(x) for x in current_values])
    plt.title(data,loc='center',pad=10, fontsize=14)
    plt.legend(loc=4)
    plt.grid(True,ls='--')

    result1 = np.where(train_loss1 == np.amin(train_loss1, axis=0))
    zipped1 = list(zip(result1[0], result1[1]))
    for cord in zipped1:
        plt.plot(cord[0]+1, train_acc1[cord], 'go')
        plt.plot(cord[0]+1, test_acc1[cord], 'b^')
```

```python
fig = plt.figure(figsize=(12,6))
plt.subplot(2,3,1)
print_data_M1('Beers',0.9)

plt.subplot(2,3,2)
print_data_M1('Flights',0.6)

plt.subplot(2,3,3)
print_data_M1('Hospital',0.6)

plt.subplot(2,3,4)
print_data_M1('Movies',0.9)

plt.subplot(2,3,5)
print_data_M1('Rayyan',0.8)

plt.subplot(2,3,6)
print_data_M1('Tax',0.8)

plt.subplots_adjust(hspace=0.5,wspace=0.3)

fig.savefig('images/M1_train_test_acc.png', dpi=100, bbox_inches='tight')
```
